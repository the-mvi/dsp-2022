## Основные теоретические положения

**Формула прямоугольников:**

$y_{n+1} = y_n + s_{n+\frac12}$, $y_0 = 0$

**Формула трапеций:**

$y_{n+1} = y_n + \frac12(s_n + s_{n+1})$, $y_0 = 0$

**Формула Симпсона:**

$y_{n+1} = y_{n-1} + \frac13(s_{n-1} + 4s_n + s_{n+1})$, $y_0 = 0$

**Формула Ньютона-Котеса:**

$y_{n+2} = y_{n-1} + \frac18(x_{n+2} + 3x_{n+1} + 3x_n + x_{n-1})$.

**Дискретный фильтр:**

Дискретный фильтр представляет собой ту или иную систему обработки дискретного сигнала, обладающую свойствами:
1. линейности – выходная реакция системы на линейную комбинацию входных сигналов равна такой же линейной комбинации ее реакций на каждый из этих сигналов отдельности;
2.  стационарности – задержка входного сигнала приводит к такой же задержке выходного сигнала без изменения его формы.

Дискретный фильтр должен обладать «памятью» т.е каждый отсчет выходного сигнала определяется в результате обработки нескольких (более одного) отсчетов входного сигнала.

Дискретный фильтр может быть задан в виде:

$y_k = b_0 x_k + b_1 x_{k-1} + ... + b_n x_{k-n} - a_1 y_{k-1} - a_2 y_{k-2} - ... - a_m y_{k-m}$.

Если $a_k = 0$, получим нерекурсивный фильтр, иначе - рекурсивный.

**Передаточная функция:**

$H(z) = {{b_0 + b_1 z^{-1} + b_2 z^{-2} + ... + b_n z^{-n}} \over {a_1 z^{-1} + a_2 z^{-2} + ... + a_m z^{-m}}}$

1. Вывести формулы передаточных функций рекурсивных фильтров, соответствующих квадратурным формулам прямоугольников, трапеций и Симпсона:

Формула прямоугольников:

$$ y_{n+1} = y_n + s_{n+\frac12}, \quad y_0 = 0 $$

Пусть $ s_n = e^{j \omega n} $ и $ y_n = H(\omega) \cdot e^{j \omega n} $, тогда:

$ H(\omega)(e^{j \omega n} \cdot e^{j \omega}) = H(\omega) \cdot e^{j \omega n} + e^{j\omega n} \cdot e^{\frac12 j\omega} $

$ H(\omega)(e^{j \omega n} e^{j \omega} - e^{j\omega n}) = e^{j\omega n} \cdot e^{\frac12 i\omega} $

$ H(\omega) \cdot e^{j \omega n} \cdot (e^{j \omega} - 1) = e^{j\omega n} \cdot e^{\frac12 i\omega} $

$ H(\omega)(e^{j\omega} - 1) = e^{\frac12 j\omega} $

$ H(\omega)(e^{\frac12 j\omega} - e^{ - \frac12 j\omega}) = 1 $

$ H(\omega) =  \frac1{e^{\frac12 j\omega} - e^{ - \frac12 j\omega}} =  \frac1{2j \sin\frac{\omega}2} $

$ \tilde H(f) = \frac1{2j\sin(\pi f)} $

Так как точное значение интеграла $ e^{j\omega t} $ равно $ \frac{e^{j\omega t}}{j\omega} $, то отношение значений:

$ \gamma = {{j \omega} \over {2j \sin{\omega \over 2}}} = {{\omega \over 2} \over \sin{\omega \over 2}} = 1 + {x^2 \over 24} + {{7x^4} \over 5760} + \dots $

$\gamma = {{\pi f} \over {\sin(\pi f)}} = 1 + {{\pi^2 f^2} \over 6} + {{7\pi^4 f^4} \over 360} + ...$

Формула трапеций:

$$ y_{n+1} = y_n + \frac12(s_n + s_{n+1}), \quad y_0 = 0 $$

Пусть $s_n = e^{j\omega n}$ и $y_n = H(\omega) \cdot e^{j\omega n}$, тогда:

$H(\omega)(e^{j\omega n} \cdot e^{j\omega}) = H(\omega) \cdot e^{j\omega n} + {{e^{j\omega n} + e^{j\omega n} \cdot e^{j\omega}} \over 2} $

$H(\omega)(e^{j\omega n} \cdot e^{j\omega}) = H(\omega) \cdot e^{j\omega n} + e^{j\omega n} {{1 + e^{j\omega}} \over 2}$

$H(\omega) \cdot e^{j\omega} = H(\omega) + {{1 + e^{j\omega}} \over 2}$

$H(\omega)(e^{j\omega} - 1) = {{1 + e^{j\omega}} \over 2}$

$H(\omega) = {{1 + e^{j\omega}} \over {2(e^{j\omega} - 1)}} = {\cos{\omega \over 2} \over {2j\sin{\omega \over 2}}}$

$\tilde{H}(f) = {{\cos(\pi f)} \over {2j\sin(\pi f)}}$

Так как точное значение интеграла $e^{j\omega t}$ равно ${e^{j\omega t} \over {j\omega}}$, то отношение значений:

$\gamma = \cos{\omega \over 2} {{\omega \over 2} \over {\sin{\omega \over 2}}} = 1 - {\omega^2 \over 12} + {\omega^4 \over 720} + \dots$

$\gamma = \cos(\pi f) {{\pi f} \over {\sin(\pi f)}} = 1 - {{\pi^2 \omega^2} \over 3} + {{\pi^4 \omega^4} \over 45} + \dots$

Формула Симпсона:

$$ y_{n+1} = y_{n-1} + \frac13(s_{n-1} + 4s_n + s_{n+1}), \quad y_0 = 0 $$

Пусть $s_n = e^{j\omega n}$ и $y_n = H(\omega) \cdot e^{j\omega n}$, тогда:

$H(\omega)(e^{j\omega n} \cdot e^{j\omega}) = H(\omega) e^{j \omega (n-1)} + \frac13(e^{j\omega (n-1)} + 4e^{j\omega n} + e^{j \omega (n+1)}) $

$H(\omega)(e^{j\omega n} \cdot e^{j\omega}) = H(\omega)e^{j\omega n}e^{-j\omega} + e^{j\omega n} {{e^{-j\omega} + 4 + e^{i \omega}} \over 3} $

$H(\omega) \cdot e^{j\omega} = H(\omega) \cdot e^{-j\omega} + {{e^{-j\omega} + 4 + e^{i \omega}} \over 3} $

$H(\omega)(e^{j\omega} - e^{-j\omega}) = {{e^{-j\omega} + 4  + e^{j\omega}} \over 3}$

$H(\omega) = {{e^{-j\omega} + 4 + e^{j\omega}} \over {3(e^{j\omega} - e^{-j\omega})}} = {{\cos\omega + 2} \over {3j\sin\omega}}$

$\tilde{H}(f) = {{\cos(2\pi f) + 2} \over {3j\sin(2 \pi f)}}$

Так как точное значение интеграла $e^{j\omega t}$ равно ${e^{j\omega t} \over {j\omega}}$, то отношение значений:

$\gamma = {{(\cos\omega + 2)j\omega} \over {3j\sin\omega}} = {{\cos\omega + 2} \over 3} \cdot {\omega \over {\sin\omega}} = 1 + {\omega^4 \over 180} + \dots$

$\gamma = {{\cos(2\pi f) + 2} \over 3} \cdot {{2\pi f} \over {\sin(2\pi f)}} = 1 + {{4\pi^4 f^4} \over 45} + ...$

Построить графики передаточных функций и графики отношения вычисляемого в результате фильтрации значения к истинному:

```python
import numpy as np
import matplotlib.pyplot as plt
import math


t = np.linspace(1e-10, 0.5, 300)
```

```python
class Rect:
    @staticmethod
    def get_H(f):
        return (1 / (2j * np.sin(math.pi*f))).imag

    @staticmethod
    def get_gamma(f):
        return math.pi * f / (np.sin(math.pi * f))
```

```python
class Trap:
    @staticmethod
    def get_H(f):
        return (np.cos(math.pi * f) / (2j * np.sin(math.pi * f))).imag

    @staticmethod
    def get_gamma(f):
        return np.cos(math.pi * f) * (math.pi * f / np.sin(f * math.pi))
```

```python
class Simpson:
    @staticmethod
    def get_H(f):
        return ((np.cos(2 * math.pi * f) + 2)/(3j * np.sin(2 * math.pi * f))).imag

    @staticmethod
    def get_gamma(f):
        return (np.cos(2 * math.pi * f) + 2) / 3
```

```python
plt.plot(t, Rect.get_H(t), label='$H_{rect}$')
plt.plot(t, Trap.get_H(t), label='$H_{trap}$')
plt.plot(t, Simpson.get_H(t), label='$H_{simpson}$')
plt.xlabel('t')
plt.ylabel('$H(t)$')
plt.legend()

x1, x2, y1, y2 = plt.axis()
plt.axis((x1, x2, -4, 1))

plt.title("Графики передаточных функций для формул прямоугольников, трапеций и Симпсона")
plt.grid()
plt.show()
```

```python
plt.plot(t, Rect.get_gamma(t), label='$\gamma_{rect}$')
plt.plot(t, Trap.get_gamma(t), label='$\gamma_{trap}$')
plt.plot(t, Simpson.get_gamma(t), label='$\gamma_{simpson}$')
plt.xlabel('t')
plt.ylabel('$\gamma(t)$')
plt.legend()

plt.title("Графики отношения, вычисляемого в результате фильтрации значения к истинному "
          "для формул прямоугольников, трапеций и Симпсона")
plt.grid()
plt.show()
```

2. Вывести формулу передаточной функции рекурсивного фильтра для интегрирования «по правилу 3/8»:

$$ y_{n+2} = y_{n-1} + {1 \over 8} (x_{n+2} + 3x_{n+1} + 3x_n + x_{n-1}). $$

Пусть $s_n = e^{j\omega n}$ и $H(\omega) \cdot e^{j\omega n}$, тогда:

$H(\omega)(e^{j\omega n} \cdot e^{2j\omega}) = H(\omega) e^{j \omega (n-1)} + {{e^{j\omega (n+2)} + 3e^{j\omega (n+1)} + 3e^{j \omega n} + e^{j\omega (n-1)}} \over 8} $

$H(\omega)(e^{j\omega n} \cdot e^{2j\omega}) = H(\omega) \cdot e^{j\omega n} e^{-j\omega} + e^{j\omega n} {{e^{2j\omega} + 3e^{j\omega} + 3 + e^{-j\omega}} \over 8}$

$H(\omega)(e^{2j\omega}) = H(\omega) \cdot e^{-j\omega} +  {{e^{2j\omega} + 3e^{j\omega} + 3 + e^{-j\omega}} \over 8}$

$H(\omega)(e^{2j\omega} - e^{-j\omega}) = {{e^{2j\omega} + 3e^{j\omega} + 3 + e^{-j\omega}} \over 8}$

$H(\omega) = {{e^{2j\omega} + 3e^{j\omega} + 3 + e^{-j\omega}} \over {8(e^{2j\omega} - e^{-j\omega})}} \cdot {{e^{-{j\omega} \over 2}} \over {e^{-{j\omega} \over 2}}}$

$H(\omega) = {{e^{{3j\omega} \over 2} + 3e^{{j\omega} \over 2} + 3e^{-{j\omega} \over 2} + e^{-{3j\omega} \over 2}} \over {8(e^{{3j\omega} \over 2} - e^{-{3j\omega} \over 2})}} = {{2\cos{{3\omega} \over 2} + 6 \cos{\omega \over 2} \over {16j\sin{{3\omega} \over 2}}}}$

$\tilde{H}(f) = {{\cos(3\pi f) + 3\cos(\pi f)} \over {8\sin(3\pi f)}}$

Так как точное значение интеграла $e^{j\omega t}$ равно ${{e^{j\omega t}} \over {j\omega}}$, то отношение значений:

$\gamma = \omega {{\cos{{3\omega} \over 2} + 3\cos{\omega \over 2}} \over {8\sin{{3\omega} \over 2}}} = {1 \over 12} (\cos{{3\omega} \over 2} + 3\cos{\omega \over 2}) \cdot {{{3\omega} \over 2} \over \sin{{3\omega} \over 2}} = {1 \over 3} + {\omega^4 \over 240} + \dots$

$\gamma = {1 \over 12}(\cos(3\pi f) + 3\cos(\pi f)){{3\pi f} \over \sin(3\pi f)} = {1 \over 3} + {{\pi^4 f^4} \over 15} + \dots$

Построить график передаточной функции и график отношения вычисляемого в результате фильтрации значения к истинному:

```python
def get_H(f):
    return ((np.cos(3 * math.pi * f) + 3 * np.cos(math.pi * f)) / (8j * np.sin(3 * math.pi * f))).imag


def get_gamma(f):
    return (1 / 12) * (np.cos(3 * math.pi * f) + 3 * np.cos(math.pi * f)) * ((3 * math.pi * f) / np.sin(3 * math.pi * f))


t = np.linspace(1e-10, 0.5, 300)
plt.plot(t, get_H(t), label='$H$', )
plt.xlabel('t')
plt.ylabel('$H(t)$')
plt.legend()

x1, x2, _, _ = plt.axis()
plt.axis((x1, x2, -30, 10))
plt.title("График передаточной функции")
plt.grid()
plt.show()
```

```python
plt.plot(t, get_gamma(t), label='$\gamma$')
plt.xlabel('t')
plt.ylabel('$\gamma(t)$')
plt.legend()

plt.title("График отношения вычисляемого в результате фильтрации значения к истинному")
plt.grid()
plt.show()
```

Анализируя графики передаточных функций рекурсивных фильтров, соответствующих квадратурным формулам прямоугольников, трапеций, Симпсона, а также формуле для интегрирования «по правилу 3/8», можно сказать, что:

1. Формула трапеций приводит к подавлению высоких частот, а формулы прямоугольников и Симпсона приводят к обратному – они усиливают высокие частоты.

2. С увеличением частоты точность интегрирования «по правилу 3/8» падает. При этом при средних частотах (около 1/3 Гц) вычисленный результат становится много больше истинного значения интеграла. При увеличении частоты выше 1/3 Гц результат интегрирования стремительно уменьшается относительно истинного значения интеграла
